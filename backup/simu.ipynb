## Backup with comparison between GM EM vs BayesianGaussianMixture for groups detection

```python
# -*- coding: utf-8 -*-
%reset -f
%matplotlib inline

import pandas as pd
import numpy as np
import pylab as pl
from tick.simulation import SimuCoxReg, SimuCoxRegWithCutPoints
from tick.preprocessing.features_binarizer import FeaturesBinarizer
from tick.inference import CoxRegression
from binacox import compute_score

from matplotlib import rc
rc('font',**{'family':'sans-serif','sans-serif':['Helvetica']})
rc('text', usetex=True)
```

```python
n_samples = 600
n_features = 2
n_cut_points = 2

seed = 3
simu = SimuCoxRegWithCutPoints(n_samples=n_samples, n_features=n_features, 
                               seed=seed, verbose=False, n_cut_points=n_cut_points,
                               shape=2, scale=.1)
X, Y, delta, cut_points, coeffs = simu.simulate()

fig = pl.figure(figsize=(12, 8))
ax = fig.add_subplot(111)
pl.scatter(*X[delta == 0].T, c=Y[delta == 0], cmap='RdBu', 
           marker="x", label="censoring")
pl.scatter(*X[delta == 1].T, c=Y[delta == 1], cmap='RdBu', 
           marker="o", label="failure")
cb = pl.colorbar()
cb.ax.tick_params(labelsize=15)

y_min, y_max = ax.get_ylim()
ax.set_ylim([y_min, y_max])
ax.plot(cut_points['0'][1:-1], np.repeat(y_min, n_cut_points), 
        'g^', ms=20, label='cut-points')
for cut_p in cut_points['0'][1:-1]:
    ax.axvline(cut_p, ls='--', c='g', alpha=.4)
x_min, x_max = ax.get_xlim()
ax.set_xlim([x_min, x_max])
ax.plot(np.repeat(x_min, n_cut_points), cut_points['1'][1:-1],
        marker=(3, 0, 3*90), c='g', ms=20)
for cut_p in cut_points['1'][1:-1]:
    ax.axhline(cut_p, ls='--', c='g', alpha=.4)

ax.tick_params(labelsize=15)
pl.tick_params(axis='x', which='both', bottom='off',
               top='off', labelbottom='off')
legend = pl.legend(loc='upper left', prop={'size': 15})
legend.legendHandles[0]._legmarker.set_markersize(10)
pl.show()

fig = pl.figure(figsize=(12, 3))
ax = fig.add_subplot(111)
ax.stem(coeffs, 'b', markerfmt='bo')
ax.set_xlim([-5, len(coeffs) + 5])
for i in range(len(cut_points)):
    ax.axvline(len(cut_points[str(i)])-1.5, c='m', ls='--', alpha=.4) 

pl.tick_params(axis='x', which='both', bottom='off',
               top='off', labelbottom='off')
pl.yticks(fontsize = 18)
ax.yaxis.labelpad = 20 
pl.show()
```

```python
# binarize features
n_cuts = 50
binarizer = FeaturesBinarizer(n_cuts=n_cuts)
X_bin = binarizer.fit_transform(X)
blocks_start = binarizer.blocks_start
blocks_length = binarizer.blocks_length
boundaries = binarizer.boundaries
```

```python
solver = 'agd' #['gd', 'agd']
learner = CoxRegression(penalty='binarsity', tol=1e-10,
                        solver=solver, verbose=False, 
                        max_iter=100, step=0.3,
                        blocks_start=blocks_start,
                        blocks_length=blocks_length)
learner._solver_obj.linesearch = False

n_folds = 10
grid_size = 30
grid_C = np.logspace(-1, 4, grid_size)
scores_cv = pd.DataFrame(columns=['ll_train', 'train_std', 'll_test', 'test_std'])
for i, C in enumerate(grid_C):
    scores_cv.loc[i] = compute_score(learner, X_bin, Y, delta, 
                                     C=C, n_folds=n_folds)
    
fig = pl.figure(figsize=(10, 8))
ax = fig.add_subplot(111)

if True:
    eps = scores_cv.train_std / np.sqrt(n_folds) * 1.96
    pl.plot(grid_C, scores_cv.ll_train, color='b')
    pl.fill_between(grid_C, 
                    scores_cv.ll_train - eps, 
                    scores_cv.ll_train + eps, 
                    color='b', alpha=0.2)

if False:
    eps = scores_cv.test_std / np.sqrt(n_folds) * 1.96
    pl.plot(grid_C, scores_cv.ll_test, color='r')
    pl.fill_between(grid_C, 
                    scores_cv.ll_test - eps, 
                    scores_cv.ll_test + eps, 
                    color='r', alpha=0.2)

idx_best = np.unravel_index(scores_cv.ll_train.argmax(), scores_cv.ll_train.shape)[0]
C_best = grid_C[idx_best]
print(C_best)

y_min, y_max = ax.get_ylim()
ax.set_ylim([y_min, y_max])
ax.plot(C_best, y_min, 'r^', ms=20, label='gamma chosen')
    
pl.title(r"CV", fontsize=18)
pl.xlabel('C', fontsize=15)
pl.xscale('log')
ax.tick_params(axis='both', which='major', direction='in',
               labelsize=17)
pl.show()
```

```python
from sklearn.model_selection import ShuffleSplit

# split data into training and test sets
test_size = .3  # proportion of data used for testing
rs = ShuffleSplit(n_splits=1, test_size=test_size, random_state=0)

for train_index, test_index in rs.split(X_bin):
        
    X_bin_test = X_bin[test_index]
    Y_test = delta[test_index]
    delta_test = delta[test_index]
    
    X_bin = X_bin[train_index]
    Y = Y[train_index]
    delta = delta[train_index]
```

```python
C_best = 30
print(C_best)
learner.C = C_best
learner.fit(X_bin, Y, delta)
print(learner.score(X_bin, Y, delta))
coeffs = learner.coeffs

fig = pl.figure(figsize=(27, 5))
ax = fig.add_subplot(111)
ax.stem(coeffs)
ax.set_xlim([-5, len(coeffs) + 5])

pl.tick_params(axis='x', which='both', bottom='off',
               top='off', labelbottom='off')
for val in blocks_start[1:]:
    ax.axvline(val - .5, c='m', ls='--') 

pl.yticks(fontsize = 18)
ax.yaxis.labelpad = 20 
pl.show()
```

```python
from sklearn.mixture import GaussianMixture as GM, BayesianGaussianMixture as BGM
import itertools

coeffs_test = coeffs[:50]
X = np.array([np.arange(len(coeffs_test)), coeffs_test]).T

jump = np.where(coeffs_test[1:] - coeffs_test[:-1] != 0)[0]
alone = np.intersect1d(jump, jump + 1)

X_train = np.delete(X, alone, axis=0)

lowest_bic = np.infty
bic = []
n_components_range = range(1, 6)
#cv_types = ['spherical', 'tied', 'diag', 'full']
cv_types = ['tied']
for cv_type in cv_types:
    for cv_type in cv_types:
        for n_components in n_components_range:
            # Fit a Gaussian mixture with EM
            gmm = GM(n_components=n_components, covariance_type=cv_type, n_init=5)
            gmm.fit(X_train)
            bic.append(gmm.bic(X_train))
            if bic[-1] < lowest_bic:
                lowest_bic = bic[-1]
                best_gmm = gmm
groups = best_gmm.predict(X)

bic = np.array(bic)
color_iter = itertools.cycle(['navy', 'turquoise', 'cornflowerblue',
                              'darkorange'])
bars = []

# Plot the BIC scores
pl.figure(figsize=(12, 7))
spl = pl.subplot(2, 1, 1)
for i, (cv_type, color) in enumerate(zip(cv_types, color_iter)):
    xpos = np.array(n_components_range) + .2 * (i - 2)
    bars.append(pl.bar(xpos, bic[i * len(n_components_range):
                                  (i + 1) * len(n_components_range)],
                        width=.2, color=color))
pl.xticks(n_components_range)
pl.ylim([bic.min() * 1.01 - .01 * bic.max(), bic.max()])
pl.title('BIC score per model')
xpos = np.mod(bic.argmin(), len(n_components_range)) + .65 +\
    .2 * np.floor(bic.argmin() / len(n_components_range))
pl.text(xpos, bic.min() * 0.97 + .03 * bic.max(), '*', fontsize=14)
spl.set_xlabel('Number of components')
spl.legend([b[0] for b in bars], cv_types)
pl.show()
```

### metrics

1) nb of cut-points detection: $\text{avg}_{\text{nb_simu}} \big\{ \text{avg}_j |\hat K_j - K^\star_j| \big\}$  
2) cut-points values estimation: $\text{avg}_{\text{nb_simu}} \big\{ \text{avg}_j \frac{1}{\hat K_j} \sum_{k=1}^{\hat K_j} \text{min}_l \{|\hat \mu_{jk} - \mu^\star_{jl}| \text{ for } l = 1, \ldots,  K^\star_j \} \big\}$
